```python
import numpy as np
import pandas as pd
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname,filename))
```

```python
ds = pd.read_csv('train.csv')
```

```python
ds.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ds.iloc[:,1:],ds.iloc[:,0],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Tree:",accuracy_score(y_test,y_pred2))
```

```python
print("Confusion matrix of Logistic Regression\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Confusion matrix of Decision Tree\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

**Logistic Regression score**

```python
print("Macro precision:", precision_score(y_test,y_pred1,average='macro'))
print("Weighted precision:", precision_score(y_test,y_pred1,average='weighted'))
precision_score(y_test,y_pred1,average=None)
```

```python
print("Macro recall:", recall_score(y_test,y_pred1,average='macro'))
print("Weighted recall:", recall_score(y_test,y_pred1,average='weighted'))
recall_score(y_test,y_pred1,average=None)
```

```python
print("Macro f1 score:", f1_score(y_test,y_pred1,average='macro'))
print("Weighted f1 score:", f1_score(y_test,y_pred1,average='weighted'))
f1_score(y_test,y_pred1,average=None)
```

**Decision Tree score**

```python
print("Macro precision:", precision_score(y_test,y_pred2,average='macro'))
print("Weighted precision:", precision_score(y_test,y_pred2,average='weighted'))
precision_score(y_test,y_pred2,average=None)
```

```python
print("Macro recall:", recall_score(y_test,y_pred2,average='macro'))
print("Weighted recall:", recall_score(y_test,y_pred2,average='weighted'))
recall_score(y_test,y_pred2,average=None)
```

```python
print("Macro f1 score:", f1_score(y_test,y_pred2,average='macro'))
print("Weighted f1 score:", f1_score(y_test,y_pred2,average='weighted'))
f1_score(y_test,y_pred2,average=None)
```
